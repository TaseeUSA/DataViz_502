#c. How many countries have hosted the Olympic Games?

```python
# Find the host countries
# answer is 23 countries
total_city_host = df['Location'].unique()

print(f"The total number of cities that have hosted the Olympic Games is {len(total_city_host)} cities\n")
print(f"These are the cities that have hosted the Olympic Games: {total_city_host}")
```

#d. How many medals usa receive?

```python
# Find usa medals
# Answer is 638 medals
usa_medals = df[df['Nationality']== "USA"].shape[0]
usa_medals_details = df[df['Nationality']== "USA"]
print(f' USA get toal {usa_medals} medals\n')
print(usa_medals_details)
```

```python
#This methods we see compare with other countries too.
print(df['Nationality'].value_counts())
```

#e. Find the total number of medals for each ntionality using group by

```python
# Group by Nationality and count their medals
nationality_medal= df.groupby('Nationality')['Medal'].count()

print("Medal totals by Nationality: ", nationality_medal)
print(len(nationality_medal))

```

```python
# Count of unique countries get the medal
nationality_count = df['Nationality'].nunique()
print(f"There are {nationality_count} unique nationalities.")
```

```python
# Check whether any country has no medals.
# Find rows where the Medal column is empty (NaN)
# Get the unique list of those country names
no_medal = df[df['Medal'].isna()]
nationality_no_medals = no_medal['Nationality'].unique()

print(f"Countries with zero medals: {nationality_no_medals}")
```
